# Visualisation des datas (Dataframe)

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer

pd.set_option('display.max_columns', 10000)
pd.set_option('display.max_rows', 200000)

chunks =  pd.read_csv("/home/sylvain/Téléchargements/openfoodfacts.csv", sep = '\t', chunksize = 1000)

df = pd.DataFrame()
for c in chunks:
    new = c
    #On vérifie si df est vide pour initialiser les colonnes
    if df.empty:
        df = new
    else:
        df = df.append(new)
    if df.shape[0] == 250000:
        break
df.head(100)
```

## On homogénéise les données textes

```python
for col in df:
    if df[col].dtypes == object:
        df[col] = df[col].str.replace('-', ' ')
        df[col] = df[col].str.replace('_', ' ')
        df[col] = df[col].str.replace(',', ' ')
        df[col] = df[col].str.replace(';', ' ')
        df[col] = df[col].str.replace(':', ' ')
        df[col] = df[col].str.lower()
        # On met NA sur les données manquantes
        mask = df[col].str.contains('unknow')
        df[col] = df[col].where(mask == False)
        mask = df[col].str.contains('to be completed')
        df[col] = df[col].where(mask == False)
```

## Suppression des doublons

```python
# sans tenir compte de la colonne code
df.drop_duplicates(subset = df.columns.values.tolist()[1:]) 
df.shape
```

```python
# sans tenir compte des colonnes Header
df.drop_duplicates(subset = df.columns.values.tolist()[7:])
df.shape
```

## On supprime les données types Header non utiles 

```python
## on supprime les colonnes header
#identification unique pour chaque ligne 
del(df['code']) 
del(df['creator'])
del(df['created_t'])
del(df['created_datetime'])
del(df['last_modified_t'])
del(df['last_modified_datetime'])

#data qui termine en tags: information redondante
for col in df.columns.values:
    if re.match(r"[\w ]*_tags", col):
        del(df[col])
    elif re.match(r"[\w _]*url", col):
        del(df[col])
        
df.shape
```

## on supprime dans un premier temps les colonnes dont les données sont manquantes à plus de 90%, pour ne pas ralentir le temps de traitements inutilement 

```python
# On supprime quand trop de NA (1st round) pour limiter les temps de traitement par la suite
for col in df:
    if df[col].count()/df.shape[0] < 0.1:
        del(df[col])
df.count()/df.shape[0]
```

## On remplace par NA les valeurs numériques erronées (data type ingredient pour 100g), on ne peut être >100g ou <0g, a l'exception de l'energie

```python
# Suppression des valeures erronées > 100g et < 0g ==> NA
for col in df:
    if col == 'energy-kcal_100g' or col == 'energy_100g':
        df[col] = df[col].where(df[col] >= 0)
    elif re.match(r"[\w -_]*100g", col):
        df[col] = df[col].where(df[col] < 100)
        df[col] = df[col].where(df[col] >= 0)

# on verifie les min et max pour les colonnes
df.min()
```

```python
df.max()
```

## on remplace les outliers par NA, restriction aux quantiles 0.1 à 0.9

```python
for col in df:
    if re.match(r"[\w -_]*100g", col):
        df[col] = df[col].where(df[col] >= df[col].quantile(0.1))
        df[col] = df[col].where(df[col] <= df[col].quantile(0.9))
            
# on visualize le % de non NA par colonne
df.count()/df.shape[0]
```

```python
# on visualize le % le ratio de valeur differentes par colonne
df.nunique()/df.shape[0]
```

## On complète les valeurs manquantes avec IterativeImputer

```python
# IterativeImputer
dft = df.copy()
dft = dft.select_dtypes(include=['float64'])

# on transfome l'ensemble des colonnes numeriques en input X
X = dft.to_numpy()
Lr = LinearRegression()
imp = IterativeImputer(estimator = Lr, max_iter = 40, tol = 1e-3, verbose = 1, imputation_order = 'arabic')
imp.fit(X)
X = imp.transform(X)
```

## On corrige les valeurs prédites et erronnées >100g et <0g

```python
#on ne recupere que les valeurs numériques
dft = pd.DataFrame(data = X, columns = dft.columns, index = df.index)

# on met 0 si prediction < 0g et 100 si prediction > 100g
for col in dft:
    if (col == 'energy-kcal_100g') or (col == 'energy_100g'):
        df[col] = dft[col].where(dft[col] > 0, 0)   
    elif re.match(r"[\w -_]*100g", col):
        df[col] = dft[col].where(dft[col] < 100, 100)
        df[col] = df[col].where(dft[col] > 0, 0)

#check 100% des datas
df.count()/df.shape[0]
```
